# Extract features

```python
import torch
import clip
import numpy as np
import os
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("RN50", device=device)
```

```python
mscoco_text_embedding = clip.tokenize(mscoco_classes).to(device)
mscoco_text_embedding.shape
```

```python
text_features = model.encode_text(mscoco_text_embedding)
text_features.shape
```

```python
text_features_cpu=[text_feature.cpu().detach().numpy() for text_feature in text_features]
```
